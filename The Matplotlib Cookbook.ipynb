# Chapter 1: Figures and Subplots


## To draw plots in matplotlib, we need something on which we can plot. Just like we use pen to draw plots on paper, on the same token, we use Python to draw plts on matplotlib figure object.

```python
# So, how can we create a figure object!?
# It's just 1 line.

fig=plt.figure()

# A good practice is to name figure object as "fig"
```

## We see that the interactive figure object gets created. But few of them are empty! Let's fill them.

```python
# plt.figure() returns a matplotlib.figure.Figure
type(fig)
```

## Now that we know where to plot, let's actually plot!

```python
fig.plot([1,2])
```

## Well, this gives error becuase, we can't  directly plot on the figure object! We should create subplots in a figure object and then plot.


```python
# To create subplots, use 

fig.add_subplot(2,2,1)

# Now this means that, we have created 2*2 ( first two parameters ) subplots.
# And the third parameter specifies the subplot which we are referring to.
```

```python
# But the problem with above code, is we did not store that in a variable. So let's quickly do it!

sub_plot_1 = fig.add_subplot(2,2,1)
```

```python
# Well, there a catchy point here. We have created 4 subplots (2*2), but we have stored only one subplot. 
# What happened to the remaining ones. Well, we can repeat the same steps above and store all the subplots.

sub_plot_2 = fig.add_subplot(2,2,2)
sub_plot_3 = fig.add_subplot(2,2,3)
sub_plot_4 = fig.add_subplot(2,2,4)
```

```python
# Now it's time to draw our first plot.

sub_plot_1.plot([1,7,2,9])
```

## Where is the output!? The output is above, scroll above and see it. But, I should get output right after the cell right? Well, it is correct, but in case of figure objects, the output is below the code where we have created figure object. We have done it in one of the previous cells i.e cell 3, so output is located there.

## So now, let's see how to get the output right after the cell

```python
fig2 = plt.figure()

fig2_sub_plot_1 = fig2.add_subplot(1,1,1)

fig2_sub_plot_1.plot([1,7,2,9])
```

# Chapter 2: Colors, markers and line styles

## Changing colors, markers and line styles aren't tough! It's very easy. All you need to do is, just add a parameter to plot function.

## Let me create another 4 subplots and then illustrate various colors, markers and line-styles.

```python

fig4 = plt.figure()

fig4_sub_plot_1 = fig4.add_subplot(2,2,1)
fig4_sub_plot_2 = fig4.add_subplot(2,2,2)
fig4_sub_plot_3 = fig4.add_subplot(2,2,3)
fig4_sub_plot_4 = fig4.add_subplot(2,2,4)

fig4_sub_plot_1.plot([1,7,2,9],'ko--')
fig4_sub_plot_2.plot(np.random.rand(10),'cv:')
fig4_sub_plot_3.plot(np.random.randn(10),'rD-.')
fig4_sub_plot_4.plot(np.random.rand(10).cumsum(),'g,-.')
```

# Chapter 3: Shapes

## As usual, let's create another 2 new subplots

```python
fig5 = plt.figure()

fig5_sub_plot_1 = fig5.add_subplot(1,2,1)
fig5_sub_plot_2 = fig5.add_subplot(1,2,2)
```

## To plot rectangle, use the below syntax!

```python
rect = plt.Rectangle((0.17,0.29),0.32,0.2)

# (0.17,0.29) is bottom left point
# 0.32 is length on X-axis
# 0.2 is length on Y-axis

fig5_sub_plot_1.add_patch(rect)
```

## Notice that we have used "add_patch" function rather than "plot" function. Basic shapes like rectangle, circle are present in matplotlib.pyplot. But, if you want all the shapes, then they are present in matplotlib.patches. Also, shapes are called as "Patches" in matplotlib.

```python
circle = plt.Circle((0.17,0.29),0.12)

# (0.17,0.29) is centre of circle
# 0.12 is radius of circle


fig5_sub_plot_2.add_patch(circle)
```

## Here is a good news! Till now, we have been choosing our styles to customize our plots. But now matplotlib does it for us. How!? Series and Data Frames are nicely integrated with matplotlib plot functions. If we call plot function with a Series or Data Frame object, matplotlib comes up with its own styling for us!
